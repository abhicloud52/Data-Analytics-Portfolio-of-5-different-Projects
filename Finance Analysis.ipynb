```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Load and explore data
def load_and_explore_finance(filename):
    df = pd.read_csv(filename)
    df.columns = df.columns.str.strip()

    print("Dataset Info:")
    print(f"Shape: {df.shape}")
    print(f"Columns: {df.columns.tolist()}")
    print("\nFirst 5 rows:")
    print(df.head())
    print("\nMissing values:")
    print(df.isnull().sum())

    return df

# Finance analysis
def analyze_finance(df):
    age_summary = df['age'].describe() if 'age' in df.columns else None
    gender_distribution = df['gender'].value_counts() if 'gender' in df.columns else None
    avenue_distribution = df['Avenue'].value_counts() if 'Avenue' in df.columns else None
    objective_distribution = df['Objective'].value_counts() if 'Objective' in df.columns else None
    source_distribution = df['Source'].value_counts() if 'Source' in df.columns else None
    monitor_distribution = df['Invest_Monitor'].value_counts() if 'Invest_Monitor' in df.columns else None

    return age_summary, gender_distribution, avenue_distribution, objective_distribution, source_distribution, monitor_distribution

# Visualization
def plot_finance_insights(gender_distribution, avenue_distribution, source_distribution, monitor_distribution):
    plt.figure(figsize=(16, 10))

    plt.subplot(2, 2, 1)
    gender_distribution.plot(kind='bar', color='skyblue')
    plt.title('Gender Distribution')

    plt.subplot(2, 2, 2)
    avenue_distribution.plot(kind='bar', color='green')
    plt.title('Preferred Investment Avenue')

    plt.subplot(2, 2, 3)
    source_distribution.plot(kind='bar', color='orange')
    plt.title('Source of Investment Information')

    plt.subplot(2, 2, 4)
    monitor_distribution.plot(kind='bar', color='purple')
    plt.title('Investment Monitoring Frequency')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    df = load_and_explore_finance(r"C:\Users\Dell\Desktop\Task2\Finance_data.csv")
    age_summary, gender_distribution, avenue_distribution, objective_distribution, source_distribution, monitor_distribution = analyze_finance(df)

    print("\nAge Summary:")
    print(age_summary)

    print("\nInvestment Objective Distribution:")
    print(objective_distribution)

    plot_finance_insights(gender_distribution, avenue_distribution, source_distribution, monitor_distribution)
```
